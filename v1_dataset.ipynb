```python
import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
from scipy.io import loadmat
from ipywidgets import interactive
```

```python
eeg_data = loadmat('data/v1_laminar.mat')
```

```python
eeg_data['csd'].shape
```

```python
srate = eeg_data['srate'][0][0]
srate
```

```python
def plot_erp(erp_signal, time_vector, channel, time_range=None):
    
    assert channel < eeg_data['csd'].shape[0], "Check your channel index"
    
    plt.figure(figsize=(10,4))
    plt.plot(time_vector, erp_signal[channel,:])
    if time_range:
        plt.xlim(time_range)
    plt.title(f"Event-Related Potential for the channel {channel}")
    plt.xlabel('Time (s)')
    plt.ylabel('Signal (microV)')
    plt.plot()
```

```python
erp = eeg_data['csd'].mean(axis=2)
times = eeg_data['timevec'][0,:]
```

```python
plot_erp(erp_signal=erp,
         time_vector=times,
         channel=6)
```

```python
plt.figure(figsize=(14,7))
plt.contourf(times, range(16), erp, cmap='RdBu_r')
plt.colorbar()
plt.yticks(range(16))
plt.xlabel('Time (s)')
plt.ylabel('Channel')
plt.title('Depth-by-time')
plt.show()
```
