```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
srate = 512
points = srate*2
times = np.arange(-points, points)/srate
```

```python
h = 1
y = np.exp(-4*np.log(2)*times**2 / h**2)
```

```python
plt.plot(times, y)
plt.axhline(0.5, linestyle='--', linewidth=1, color='red')
plt.axvline(0+h/2, linestyle='--', linewidth=1, color='red')
plt.axvline(0-h/2, linestyle='--', linewidth=1, color='red')
plt.show()
```

```python
pnts = 10000
srate = 1024
times = np.arange(pnts)/srate
```

```python
# chrips

## bipolar chirp
freqmod = np.linspace(5, 15, pnts)
```

```python
signal = np.sin(2*np.pi*(np.cumsum(freqmod)/srate + times))
```

```python
plt.figure(figsize=(15,4))
plt.plot(times, signal)
plt.show()
```

```python
# multipolar chirp
k = 10 # poles for frequences

x = np.linspace(0, 10, k)
y = np.random.randint(1, 20, k)
xvals = times
freqmod = np.interp(xvals, x, y)

plt.plot(x, y, 'o', xvals, freqmod, '-')
plt.show()
```

```python
signal = np.sin(2*np.pi*(np.cumsum(freqmod)/srate + times))

plt.figure(figsize=(15,4))
plt.plot(times, signal)
plt.show()
```
