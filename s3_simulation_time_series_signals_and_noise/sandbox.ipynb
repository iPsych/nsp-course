```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
srate = 512
points = srate*2
times = np.arange(-points, points)/srate
```

```python
h = 1
y = np.exp(-4*np.log(2)*times**2 / h**2)
```

```python
plt.plot(times, y)
plt.axhline(0.5, linestyle='--', linewidth=1, color='red')
plt.axvline(0+h/2, linestyle='--', linewidth=1, color='red')
plt.axvline(0-h/2, linestyle='--', linewidth=1, color='red')
plt.show()
```

```python
pnts = 10000
srate = 1024
times = np.arange(pnts)/srate
```

```python
# chrips

## bipolar chirp
freqmod = np.linspace(5, 15, pnts)
```

```python
signal = np.sin(2*np.pi*(np.cumsum(freqmod)/srate + times))
```

```python
plt.figure(figsize=(15,4))
plt.plot(times, signal)
plt.show()
```

```python
# multipolar chirp
k = 10 # poles for frequences

x = np.linspace(0, 10, k)
y = np.random.randint(1, 20, k)
xvals = times
freqmod = np.interp(xvals, x, y)

plt.plot(x, y, 'o', xvals, freqmod, '-')
plt.show()
```

```python
signal = np.sin(2*np.pi*(np.cumsum(freqmod)/srate + times))

plt.figure(figsize=(15,4))
plt.plot(times, signal)
plt.show()
```

```python
pnts = 4567
srate = 987

peakfreq = 30
fwhm = 5
```

```python
hz = np.linspace(0, srate, pnts)

s = fwhm * (2*np.pi - 1)/(4*np.pi)
x = hz - peakfreq
fg = np.exp(-0.5*(x/s)**2)
```

```python
# Fourier coefficients
fc = np.random.rand(pnts) * np.exp(complex(imag=1)*2*np.pi*np.random.rand(pnts))
fc = fc*fg

signal = 2*np.fft.ifft(fc).real
```

```python
plt.figure(figsize=(14,5))
plt.plot(hz, np.abs(fc))
plt.xlim([0, 3*peakfreq])
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(signal)
plt.show()
```

```python
pnts = 4000
srate = 1000
time = np.arange(pnts)/srate - 1
```

```python
# gaussian parameters
peaktime = 1
fwhm = 0.4

# sine wave parameters 
sinefreq = 7
```

```python
gaus = np.exp(-4*np.log(2) * (time-peaktime)**2/fwhm**2)
cosw = np.cos(2*np.pi*sinefreq*time + 2*np.pi*np.random.rand(1)) #non-phase locked

signal = gaus*cosw
```

```python
plt.figure(figsize=(14,5))
plt.plot(time, signal)
plt.show()
```
