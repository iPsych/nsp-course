```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy import signal
```

```python
eeg_data = loadmat('../data/v1_laminar.mat')
srate = eeg_data['srate'][0][0]
timevec = eeg_data['timevec']
```

```python
eeg_data['csd'].shape
```

```python
data = eeg_data['csd'][6,:,:]
```

```python
# cut-off frequency
lowcut = 20 # Hz
```

```python
plt.figure(figsize=(15,7))
plt.imshow(data.T, aspect='auto', cmap='RdBu_r')
plt.xlabel('Time')
plt.ylabel('Trial')
plt.colorbar()
plt.show()
```

```python
# %% create and inspect the filter kernel

# % filter order
filtord = int(18*(lowcut*1000/srate))
filtord
```

```python
# % create filter
filtkern = signal.firwin(filtord, lowcut/(srate/2), window="hamming", pass_zero='lowpass')
```

```python
# % inspect the filter kernel
plt.plot(filtkern)
plt.xlabel('Time (s)')
plt.title('Time domain')
plt.show()
```

```python
hz = np.linspace(0, srate, filtkern.shape[0])

plt.plot(hz, (np.fft.fft(filtkern)**2).real)
plt.xlim([0, lowcut*3])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.title('Frequency domain')
plt.show()
```

```python
# %% option 1: filter the ERP

# % extract ERP
erp1 = np.mean(data,1)

# % apply filter
erp1_filtered = signal.filtfilt(filtkern,1,erp1)
```

```python
plt.figure(figsize=(10,5))
plt.plot(erp1)
plt.plot(erp1_filtered, color='red')
plt.show()
```

```python
# %% option 2: filter the single trials

erp2 = np.zeros(timevec.shape[1])
erp2.shape
```

```python
for trial in range(data.shape[1]):
    erp2 = erp2 + signal.filtfilt(filtkern,1,data[:, trial])

# % complete the averaging
erp2 = erp2/trial
```

```python
plt.figure(figsize=(10,5))
plt.plot(erp1)
plt.plot(erp2, color='red')
plt.show()
```
