```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import loadmat
from scipy import signal
```

```python
eeg_data = loadmat('../data/sampleEEGdata.mat')
data = eeg_data['EEG']['data'][0][0]
times = eeg_data['EEG']['times'][0][0][0]
data.shape
```

```python
# butterfly plot
plt.figure(figsize=(12,5))
for ch in range(data.shape[0]):
    plt.plot(times, data[ch,:,:].mean(axis=1))
plt.xlim([-500, 1200])
plt.xlabel('Time (s)')
plt.ylabel('Voltage (muV)')
plt.title('Earlobe reference')
plt.show()
```

```python
car = data - data.mean(axis=0)
```

```python
plt.figure(figsize=(12,5))
for ch in range(data.shape[0]):
    plt.plot(times, car[ch,:,:].mean(axis=1))
plt.xlim([-500, 1200])
plt.xlabel('Time (s)')
plt.ylabel('Voltage (muV)')
plt.title('Average reference')
plt.show()
```

```python
var_ts_ER = data.mean(axis=2).var(axis=0)
var_ts_AR = car.mean(axis=2).var(axis=0)
```

```python
plt.figure(figsize=(12,5))
plt.plot(times, var_ts_ER)
plt.plot(times, var_ts_AR)
plt.xlim([-500, 1200])
plt.xlabel('Time (s)')
# plt.ylabel('Voltage (muV)')
plt.title('Topographical Variance Time series')
plt.show()
```
